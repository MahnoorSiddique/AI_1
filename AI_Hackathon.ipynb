```python
!pip install python-pptx
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN
from pptx.enum.shapes import MSO_SHAPE
from google.colab import files

def create_final_pitch_deck():
    prs = Presentation()
    prs.slide_width, prs.slide_height = Inches(13.33), Inches(7.5)

    def apply_dark_theme(slide):
        bg = slide.shapes.add_shape(MSO_SHAPE.RECTANGLE, 0, 0, prs.slide_width, prs.slide_height)
        bg.fill.solid()
        bg.fill.fore_color.rgb = RGBColor(13, 13, 20) # Midnight Blue/Black
        bg.line.fill.background()

    def add_neon_title(slide, text):
        title = slide.shapes.add_textbox(Inches(0.5), Inches(0.4), Inches(12), Inches(1))
        p = title.text_frame.paragraphs[0]
        p.text = text
        p.font.size, p.font.bold = Pt(48), True
        p.font.color.rgb = RGBColor(0, 255, 255) # Electric Cyan

    # --- SLIDE 1: THE BRAND ---
    slide = prs.slides.add_slide(prs.slide_layouts[6]); apply_dark_theme(slide)
    t = slide.shapes.add_textbox(Inches(1.5), Inches(2.5), Inches(10), Inches(2))
    p = t.text_frame.paragraphs[0]
    p.text = "NEURAL RESURRECT"
    p.font.size, p.font.bold, p.alignment = Pt(88), True, PP_ALIGN.CENTER
    p.font.color.rgb = RGBColor(255, 255, 255)
    sub = slide.shapes.add_textbox(Inches(1.5), Inches(4.2), Inches(10), Inches(1))
    sp = sub.text_frame.paragraphs[0]
    sp.text = "Ending the 'Zombie State' of Passive Education | AI Hackathon 2024"
    sp.font.size, sp.font.color.rgb, sp.alignment = Pt(24), RGBColor(0, 200, 200), PP_ALIGN.CENTER

    # --- SLIDE 2: THE DATASET ANALYSIS ---
    slide = prs.slides.add_slide(prs.slide_layouts[6]); apply_dark_theme(slide)
    add_neon_title(slide, "A MULTI-DIMENSIONAL DATA ENGINE")

    # Specific counts extracted from your code
    data_points = [
        ("MATH (70)", "Logic, Arithmetic, Series Analysis"),
        ("SCIENCE (80)", "Physics, Bio, Astronomy, Chemistry"),
        ("LANGUAGES (70)", "French, Arabic, Urdu, Spanish"),
        ("ISLAMIC (80)", "Theology, Pillars, History")
    ]

    for i, (cat, desc) in enumerate(data_points):
        box = slide.shapes.add_shape(MSO_SHAPE.RECTANGLE, Inches(0.5 + (i*3.2)), Inches(2), Inches(3), Inches(4))
        box.fill.solid(); box.fill.fore_color.rgb = RGBColor(30, 30, 45)
        box.line.color.rgb = RGBColor(0, 255, 255)
        tf = box.text_frame
        tf.paragraphs[0].text = f"{cat}\n\n{desc}"
        tf.paragraphs[0].font.size = Pt(22)

    # --- SLIDE 3: THE REINFORCEMENT LOGIC ---
    slide = prs.slides.add_slide(prs.slide_layouts[6]); apply_dark_theme(slide)
    add_neon_title(slide, "TECHNICALITY: THE MASTERY LOOP")

    code_text = (
        "while True:\n"
        "    # Step 1: Pseudorandom Selection from Dictionary\n"
        "    q = random.choice(category_pool)\n"
        "    \n"
        "    # Step 2: Persistent State Check\n"
        "    if user_input != q['answer']:\n"
        "        print('ðŸ’€ WRONG! Try again.')\n"
        "        # Loop does not break: Forced Retrieval triggered\n"
        "    else:\n"
        "        print('âœ¨ CORRECT!')\n"
        "        break"
    )

    code_box = slide.shapes.add_textbox(Inches(1), Inches(1.8), Inches(11), Inches(5))
    p = code_box.text_frame.paragraphs[0]
    p.text = code_text
    p.font.name, p.font.size, p.font.color.rgb = "Consolas", Pt(22), RGBColor(50, 255, 50)

    # --- SLIDE 4: USER PROGRESSION & UX ---
    slide = prs.slides.add_slide(prs.slide_layouts[6]); apply_dark_theme(slide)
    add_neon_title(slide, "USER EXPERIENCE: PROGRESSION")
    content = slide.shapes.add_textbox(Inches(1), Inches(2), Inches(11), Inches(4)).text_frame
    pts = [
        "â€¢ Difficulty Normalization: 40-question 'Beginner' sprints.",
        "â€¢ State Management: Real-time score tracking [X/40].",
        "â€¢ Clean UI: Simple A/B/C interface to reduce cognitive load.",
        "â€¢ Safe Exit Protocols: Error handling for non-standard inputs."
    ]
    for pt in pts:
        p = content.add_paragraph()
        p.text = pt
        p.font.size, p.font.color.rgb = Pt(32), RGBColor(220, 220, 220)

    # --- SLIDE 5: SCALABILITY (The AI Pitch) ---
    slide = prs.slides.add_slide(prs.slide_layouts[6]); apply_dark_theme(slide)
    add_neon_title(slide, "FUTURE ROADMAP: GENERATIVE AI")
    content = slide.shapes.add_textbox(Inches(1), Inches(2), Inches(11), Inches(4)).text_frame
    pts = [
        "â€¢ LLM Content Generation: Using Gemini to create infinite, unique question nodes.",
        "â€¢ Adaptive Difficulty ML: Scaling math complexity based on response time latency.",
        "â€¢ Natural Language Interface: Replacing A/B/C with conversational evaluation."
    ]
    for pt in pts:
        p = content.add_paragraph()
        p.text = pt
        p.font.size, p.font.color.rgb = Pt(32), RGBColor(255, 255, 255)

    name = "Neural_Resurrect_Final_Hackathon.pptx"
    prs.save(name)
    files.download(name)

create_final_pitch_deck()
```
